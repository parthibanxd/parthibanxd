```python
import random
import time

THRESHOLD_PM25 = 75
CRITICAL_DO = 4.0

SENSOR_COUNT_PM = 3
DO_SENSOR_ID = "S01"

ITER = 20

print("----- PM2.5 REAL-TIME MONITORING -----")
for i in range(ITER):
    print(f"\nTime step {i+1}")
    for sid in range(1, SENSOR_COUNT_PM + 1):
        pm = random.randint(10, 100)
        print(f"PM2.5_Sensor_{sid}: {pm}", end='')
        print(" <-- ALERT" if pm > THRESHOLD_PM25 else "")
    time.sleep(1)

print("\n----- DISSOLVED OXYGEN (DO) MONITORING -----")
for _ in range(ITER):
    value = round(random.uniform(3.0, 7.0), 2)
    ts = time.strftime("%H:%M:%S")
    print(f"[ALERT] DO:{value}, Time:{ts}"
          if value < CRITICAL_DO
          else f"Sensor:{DO_SENSOR_ID}, DO:{value}, Time:{ts}")
    time.sleep(1)
```
